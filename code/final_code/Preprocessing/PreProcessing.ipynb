```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../../data/clean_data/salaries_descriptions_locations_gramcounts.csv')
```

```python
country_encoding = pd.get_dummies(df.Country, prefix='Country')
```

```python
df = df.merge(country_encoding,left_index=True,right_index=True)
```

```python
df = df.drop(['Country','Description','company','base_pay','Glassdoor Salary Estimate'],axis=1)
```

```python
remote_ind = []
for index,row in df.iterrows():
    if row['company_location'] == 'Remote':
        remote_ind.append(1)
    else:
        remote_ind.append(0)
```

```python
df['Remote'] = remote_ind
df = df.drop(['company_location'],axis=1)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
def position_map(pos_hierarchy,df):
    pos_mapping = {'Data Scientist Intern':pos_hierarchy[0],
               'Junior Data Scientist':pos_hierarchy[1],
               'Data Scientist':pos_hierarchy[2],
               'Senior Data Scientist':pos_hierarchy[3],
               'Lead Data Scientist':pos_hierarchy[4]}
    
    df_mapped = df.replace({'Position_clean':pos_mapping})
    return df_mapped
```

```python
pos_ranks = [1,2,3,4,5]
df = position_map(pos_ranks,df)
```

```python
df
```

```python
normalized_df=(df-df.mean())/df.std()
```

```python
mask = np.random.rand(len(normalized_df))<0.8
train_norm = normalized_df[mask]
test_norm = normalized_df[~mask]

train = df[mask]
test = df[~mask]
```

```python
train.to_csv('../../../data/clean_data/train.csv')
test.to_csv('../../../data/clean_data/test.csv')

train_norm.to_csv('../../../data/clean_data/train_normalized.csv')
test_norm.to_csv('../../../data/clean_data/test_normalized.csv')
```
