```python
import pandas as pd
from geopy import geocoders
import time
```

```python
gn = geocoders.GeoNames(username='sashananda')
```

```python
def convert_city_to_lat_long(city):
    lat = float('nan')
    long = float('nan')
    if (type(city) != float):
        print(city)
        place, (lat, long) = gn.geocode(city)
    return (lat, long)
```

```python
def add_lat_long_columns(df, location_col_name):
    location_col = df[location_col_name]
    locations = {}
    lats = []
    longs = []
    for location in location_col:
        time.sleep(4)
        print(location)
        try:
            lat, long = locations[location]
        except:
            lat, long = convert_city_to_lat_long(location)
            locations[location] = (lat, long)
        lats.append(lat)
        longs.append(long)
#     lat_long = location_col.map(convert_city_to_lat_long)
#     df['company_lat'] = lat_long.map(lambda x: x[0])
#     df['company_long'] = lat_long.map(lambda x: x[1])    
    df['company_lat'] = lats
    df['company_long'] = longs
    return df
```

```python
company_info_ca = pd.read_csv('../../../data/clean_data/company_info_ca.csv', index_col=0)
company_info_us = pd.read_csv('../../../data/clean_data/company_info_us.csv', index_col=0)
company_info_uk = pd.read_csv('../../../data/clean_data/company_info_uk.csv', index_col=0)
```

```python
company_info_ca['company_location'] = company_info_ca['company_location'].str.replace('Mc Lean', 'McLean')
company_info_ca = add_lat_long_columns(company_info_ca, 'company_location')
```

```python
company_info_ca.to_csv('../../../data/clean_data/company_info_ca.csv')
```
