```python
import pandas as pd
```

```python
df_loc_ca = pd.read_csv('../../../data/clean_data/company_info_ca.csv')
df_loc_uk = pd.read_csv('../../../data/clean_data/company_info_uk.csv')
df_loc_us = pd.read_csv('../../../data/clean_data/company_info_us.csv')
```

```python
df_loc_ca = df_loc_ca[['company_location','company_lat','company_long']]
df_loc_uk = df_loc_uk[['company_location','company_lat','company_long']]
df_loc_us = df_loc_us[['company_location','company_lat','company_long']]

df_loc = pd.concat([df_loc_ca,df_loc_uk,df_loc_us])
```

```python
df = pd.read_csv('../../../data/clean_data/salaries_and_descriptions.csv')
```
